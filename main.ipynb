```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# datset: https://www.kaggle.com/datasets/fratzcan/usa-house-prices

data = pd.read_csv("dataset.csv")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data = data[data['price'] > 0]
data.describe()
```

```python
data = data.drop(columns=['country', 'street'])
data.info()
```

```python
# turing date column into just the year
data['sale_year'] = pd.to_datetime(data['date']).dt.year
data = data.drop(columns=['date'])
data.head()
```

```python
data.duplicated().sum()
```

```python
print((data['bedrooms'] == 0).sum())
print((data['bathrooms'] == 0).sum())
```

```python
data[data['bedrooms'] == 0]
```

```python
data[data['bathrooms'] == 0]
```

```python
data = data[(data['bathrooms'] > 0)]
print((data['bedrooms'] == 0).sum())
print((data['bathrooms'] == 0).sum())
```

```python
# looking for columns to feature engineer
data.head()
```

```python
data['statezip'].nunique()
```

```python
# instead of using year built and sale year, we can just calucate the age of the house at the time of the sale
data['age_at_sale'] = data['sale_year'] - data['yr_built']

# we can also change the renovation year to be the years since the renovation
data['years_since_renovation'] = data['sale_year'] - np.where(data['yr_renovated'] == 0, data['yr_built'], data['yr_renovated'])

# above ground ratio tells us how much of the house is above ground (not basement)
data['above_ground_ratio'] = data['sqft_above'] / data['sqft_living']

# drop the columns we just feature engineered
data = data.drop(columns=['yr_built', 'yr_renovated'])

# one-hot encode the statezip column
data = pd.get_dummies(data, columns=['city', 'statezip'], drop_first=True, dtype=int)
```

```python
data.head()
```

```python
data.columns.tolist()
```

```python
# reorder columns
reorder_cols = ['price', 
                 'age_at_sale', 'years_since_renovation', 'above_ground_ratio',
                 'sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement',
                 'bedrooms', 'bathrooms', 'floors',
                 'waterfront', 'view', 'condition', 'sale_year']
other_cols = [col for col in data.columns if col not in reorder_cols]
data = data[reorder_cols + other_cols]

data.head()
```

```python
# split data into train (80%) and test (20%) sets
from sklearn.model_selection import train_test_split

X = data.drop(columns=['price'])
y = data['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# scale features 
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# build and train the model
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)

plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs Predicted Prices')
plt.show()
```

```python
# predict on the test set
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
y_pred = model.predict(X_test_scaled)

print("R^2: ", r2_score(y_test, y_pred))
print("Mean Absolute Error: ", mean_absolute_error(y_test, y_pred))
print("Mean Squared Error: ", mean_squared_error(y_test, y_pred))

# R^2: how much price variation is explained by model (1 is perfect 0 is useless)
# Mean Absolute Error: average dollar amount prediction is off by
# Mean Squared Error: like MAE but penalizes larger errors more
```

```python
# shap analysis - explains the output of machine learning models
import shap

explainer = shap.LinearExplainer(model, X_train_scaled)
shap_values = explainer.shap_values(X_test_scaled)

shap.summary_plot(shap_values, X_test_scaled, feature_names=X.columns, plot_type="bar")


```

SHAP Analysis Findings

- Location is a key driver; for example, a house being in Seattle has the highest impact in changing price.
- Size features inflence. As expected the larger the size of the house the higher the price it would be.
- Floor and bathrooms seem to have a higher influence over bedrooms (number of).
- Some wealthy zip codes stand out: WA 98112 (Capitol Hill/Madison Park), WA 98004 (downtown Bellevue), and WA 98199 (Magnolia)
- Condition and view have moderate impact. I thought these would sneakily be on the top.
- 'above_ground_ratio' ranked #2, showing that sometimes engineered features can be more predictive than raw columns.
